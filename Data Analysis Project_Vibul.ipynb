```python
#NPS Score Calculation

df1=pd.DataFrame(data)
selected_columns = ['Inflight wifi service','Departure/Arrival time convenient','Ease of Online booking','Gate location','Food and drink','Online boarding','Seat comfort','Inflight entertainment','On-board service','Leg room service','Baggage handling','Checkin service','Inflight service','Cleanliness']

df1['sum'] = df1[selected_columns].sum(axis=1) #Calculates the sum of ratings

Total_score = len(selected_columns) #calculates the total rating

df1['Rating Percentage'] = (df1['sum']/(Total_score*5)) #Rating in percentage

#Defining rating range

detractor_range = (0.0, 0.60)
passive_range = (0.61, 0.85)
promoter_range = (0.86, 1)

df1['Score Category'] = ''

for index, row in df1.iterrows():
    score = row['Rating Percentage']
    if score >= detractor_range[0] and score <= detractor_range[1]:
        df1.at[index, 'Score Category'] = 'Detractor'
    elif score >= passive_range[0] and score <= passive_range[1]:
        df1.at[index, 'Score Category'] = 'Passive'
    elif score >= promoter_range[0] and score <= promoter_range[1]:
        df1.at[index, 'Score Category'] = 'Promoter'
        
promoter_percentage = (df1['Score Category'].value_counts()['Promoter'] / len(df1)) * 100
Passive_percentage = (df1['Score Category'].value_counts()['Passive'] / len(df1)) * 100
Detractor_percentage = (df1['Score Category'].value_counts()['Detractor'] / len(df1)) * 100

#NPS calculation

NPS = promoter_percentage - Detractor_percentage
print("The NPS is :", NPS)
```
